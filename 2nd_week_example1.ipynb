```python
from requests import get  
from PIL import Image
import gzip, os, sys, matplotlib
import numpy as np
from struct import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def download(url, file_name = None):
    if not file_name:
        file_name = url.split('/')[-1]
    
    with open(file_name, "wb") as file:
        response = get(url)
        file.write(response.content)        
        
if __name__ == '__main__':
    url = "http://yann.lecun.com/exdb/mnist/train-images-idx3-ubyte.gz"
    download(url)
    
    a="C:\\Users\\USER\\Jupyter Notebook\\RP_Lab _DL\\"
    b=a+url[33:]
    
    input = gzip.GzipFile(b, 'rb')
    s = input.read()
    input.close()
    output = open(a+url[33:-3], 'wb')
    output.write(s)
    
if __name__ == '__main__':
    url = "http://yann.lecun.com/exdb/mnist/train-labels-idx1-ubyte.gz"
    download(url)
    a="C:\\Users\\USER\\Jupyter Notebook\\RP_Lab _DL\\"
    b=a+url[33:]
    
    input = gzip.GzipFile(b, 'rb')
    s = input.read()
    input.close()
    output = open(a+url[33:-3], 'wb')
    output.write(s)
    
    
if __name__ == '__main__':
    url = "http://yann.lecun.com/exdb/mnist/t10k-images-idx3-ubyte.gz"
    download(url)
    a="C:\\Users\\USER\\Jupyter Notebook\\RP_Lab _DL\\"
    b=a+url[33:]
    
    input = gzip.GzipFile(b, 'rb')
    s = input.read()
    input.close()
    output = open(a+url[33:-3], 'wb')
    output.write(s)
    
    
if __name__ == '__main__':
    url = "http://yann.lecun.com/exdb/mnist/t10k-labels-idx1-ubyte.gz"
    download(url)
    a="C:\\Users\\USER\\Jupyter Notebook\\RP_Lab _DL\\"
    b=a+url[33:]
    
    input = gzip.GzipFile(b, 'rb')
    s = input.read()
    input.close()
    output = open(a+url[33:-3], 'wb')
    output.write(s)

print("Download and extract complete\n")
##############################여기까지 다운로드 & 압축풀기#################################

#남은 것 : 파싱을 이용한 바이트 --> 이미지 변환
#          변환된 이미지 해당 디렉토리에 저장시키기

#디렉토리 생성
dir = 'C:/Users/USER/Jupyter Notebook/RP_Lab _DL/mnist_dataset/'
if not os.path.exists(dir):
    os.mkdir(dir)
dir = 'C:/Users/USER/Jupyter Notebook/RP_Lab _DL/mnist_dataset/train/'
if not os.path.exists(dir):
    os.mkdir(dir)
dir = 'C:/Users/USER/Jupyter Notebook/RP_Lab _DL/mnist_dataset/test/'
if not os.path.exists(dir):
    os.mkdir(dir)    
for i in range(10):
    labelnumberdir1 = 'C:/Users/USER/Jupyter Notebook/RP_Lab _DL/mnist_dataset/train/' + str(i)+'/'
    if not os.path.exists(labelnumberdir1):
        os.mkdir(labelnumberdir1)
    labelnumberdir2 = 'C:/Users/USER/Jupyter Notebook/RP_Lab _DL/mnist_dataset/test/' + str(i)+'/'
    if not os.path.exists(labelnumberdir2):
        os.mkdir(labelnumberdir2)

#파일 읽기 
fp1_image = open('train-images-idx3-ubyte','rb')
fp1_label = open('train-labels-idx1-ubyte','rb')
fp2_image = open('t10k-images-idx3-ubyte','rb')
fp2_label = open('t10k-labels-idx1-ubyte','rb')

#사용할 변수 초기화
img = np.zeros((28,28))         #이미지가 저장될 부분 (28x28)
lbl = [ [],[],[],[],[],[],[],[],[],[] ] #숫자별로 저장 (0 ~ 9)
d = 0
l = 0
index=0 

s = fp1_image.read(16)    #read first 16byte
l = fp1_label.read(8)        #read first  8byte
s_t = fp2_image.read(16)
l_t = fp2_label.read(8)

#숫자 데이터를 읽어서 해당하는 데이터를 지정하고 출력 

#read mnist and show number
for i in range(10):
    k = i
    
    while True:    
        s = fp1_image.read(784) #784바이트씩 읽음
        l = fp1_label.read(1) #1바이트씩 읽음
        s_t = fp2_image.read(784)
        l_t = fp2_label.read(1) 
        
        if not s:
            break;
        if not l:
            break;

        index = int(l[0]) 
        index_t = int(l_t[0])
        
        #unpack
        img = np.reshape( unpack(len(s)*'B',s), (28,28))       #unpack(format, string)
        img_t = np.reshape( unpack(len(s_t)*'B',s_t), (28,28)) 
        lbl[index].append(img) #각 숫자영역별로 해당이미지를 추가
        lbl[index].append(img_t)
                
        #print(img)

        plt.imshow(img, cmap = cm.binary) #binary형태의 이미지 설정
        fig = plt.gcf()
        savename = str(k)+"mnist.jpg"
        for o in range(0, 10):
            if index == o:
                fig.savefig('C:/Users/USER/Jupyter Notebook/RP_Lab _DL/mnist_dataset/train/'+str(o)+'/'+savename)
                
        plt.imshow(img_t,cmap = cm.binary)
        fig2 = plt.gcf()    
        for p in range(0, 10):
            if index_t == p:
                fig2.savefig('C:/Users/USER/Jupyter Notebook/RP_Lab _DL/mnist_dataset/test/'+str(p)+'/'+savename)
                
        k = k + 1    
        
        #print(np.shape(lbl)) #label별로 잘 지정됬는지 확인

print("save done")
```
